```python
import transformers
import torch

model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",token = access_token
)

```

```python
messages = [
    {"role": "system", "content": "You are a expert agriculture domain chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "What are the ways to stop incests in paddy?"},
]

outputs = pipeline(
    messages,
    max_new_tokens=256,
)
#print(outputs[0]["generated_text"][-1])
```

```python
from pprint import pprint
pprint(outputs[0]['generated_text'][-1])
```

```python
messages = [
    {"role": "system", "content": "You are a expert customer support agriculture domain chatbot who is expert in Tamil and Malayalam language!"},
    {"role": "user", "content": "What are the ways to stop incests in paddy?"},
]

outputs = pipeline(
    messages,
    max_new_tokens=300
)
pprint(outputs[0]["generated_text"][-1])
```
